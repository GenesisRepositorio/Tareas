```python
from tkinter import ttk
import tkinter as tk
from tkinter import *
import sqlite3
import random  
from tkinter import messagebox as mb
import os
import random
from tkinter import scrolledtext
from tkinter import PhotoImage
from PIL import Image,ImageTk
from tkPDFViewer import tkPDFViewer as pdf
import fitz

class sesion:#nombre de la clase
    
    def __init__(self, res): #constructor

        self.frame = Frame(res)
        self.lblS=Label(self.frame,text="Login")
        self.frame.pack(side = tk.LEFT)
        self.lblS.pack(side = tk.TOP)
        #self.icon = PhotoImage(file='cerebro.png')#icono
        #self.frame.iconphoto(True, self.icon)
        #self.frame.resizable(0, 0)
        #self.frame['bg'] = 'skyblue'#color de la pestaña 

        
        style = ttk.Style()
        style.map("C.TButton",foreground=[('pressed', 'red'), ('active', 'blue')],background=[('pressed', '!disabled', 'black'), ('active', 'white')])

        
        self.image1 = tk.PhotoImage(file="correo-electronico-vacio.png")
        self.lb=Label(self.frame , image=self.image1).pack(side = tk.LEFT)
        
        CorreoLabel=Label(self.frame , text = 'Correo:')
        CorreoLabel.pack(side = tk.LEFT)
        self.correo = Entry(self.frame , fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.correo.focus()
        self.correo.pack(side = tk.LEFT)
        
        self.image2 = tk.PhotoImage(file="candado.png")
        Label(self.frame , image=self.image2).pack(side = tk.LEFT)
        Label(self.frame ,text= 'Contraseña' ).pack(side = tk.LEFT)
        self.cont = Entry(self.frame , fg="#1034e8", font=("bold", 11), show='*', highlightthickness=2)
        self.cont.focus()
        self.cont.pack(side = tk.LEFT)
        

        ttk.Button(self.frame,text="Iniciar sesión", command=self.login,style="C.TButton").pack(side = tk.LEFT)
        ttk.Button(self.frame,text="Registrarse",command=self.registro,style="C.TButton").pack(side = tk.BOTTOM)
        ttk.Button(self.frame,text="Contraseña olvidada",command=self.contraseña,style="C.TButton").pack(side=tk.BOTTOM)
        

        self.img = Image.open('maestr2.png')
        self.img = self.img.resize((200, 200), Image.ANTIALIAS) # Redimension (Alto, Ancho)
        self.img = ImageTk.PhotoImage(self.img)
        widget = Label(self.frame, image=self.img)
        widget.pack(side = tk.BOTTOM)
        
    def login(self):
        db = sqlite3.connect( 'db_math.db')
        c = db.cursor()
        usuario = self.correo.get()
        contr = self.cont.get()
        c.execute('SELECT * FROM ALUMNO WHERE correo = ? AND contraseña = ?', (usuario, contr))
        if c.fetchall():
            mb.showinfo(title = "Login correcto", message = "Usuario y contraseña correctos")
            self.nivel()
        else:
            mb.showerror(title = "Login incorrecto", message = "Usuario o contraseña incorrecta")  
        c.close()
          
    
    def registro(self):

        self.rest = Tk()
        self.rest.title("Registrarse")
        self.rest['bg'] = 'skyblue'
        self.rest.geometry('300x300+500+200')
        
        frame = LabelFrame(self.rest, text = 'Registro',bg='#F3FFD6')
        frame.pack()
        
        db = sqlite3.connect( 'db_math.db')
        c = db.cursor()
        
        self.rut_A = StringVar()
        self.establ= StringVar()
        self.comuna= StringVar()
        self.region= StringVar()
        self.nom_A = StringVar()
        self.edad = IntVar()
        self.contraseña = StringVar()
        self.correo = StringVar()
        
        
        
        Label(frame, text = 'Rut alumno:').grid(row =4, column = 0)       
        self.rut_A = Entry(frame,textvariable=self.rut_A, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.rut_A.focus()
        self.rut_A.grid(row = 4, column = 1)
        
        query3='SELECT nombre_establec FROM ESTABLECIMIENTO'
        c.execute(query3)
        milista3=[]
        for row in c.fetchall():
            milista3.append(row[0])
        Label(frame, text = 'Establecimiento :').grid(row = 1, column = 0) 
        self.establ = ttk.Combobox(frame,textvariable=self.establ,values=milista3)
        self.establ.focus()
        self.establ.grid(row = 1, column =1) 
        self.establ.current(0)
                
        query1='SELECT nombre_com FROM COMUNA'
        c.execute(query1)
        milista1=[]
        for row in c.fetchall():
            milista1.append(row[0])
        Label(frame, text = 'Comuna :').grid(row = 2, column = 0) 
        self.comuna = ttk.Combobox(frame,textvariable=self.comuna,values=milista1)
        self.comuna.focus()
        self.comuna.grid(row = 2, column =1) 
        self.comuna.current(0)
        
        query2='SELECT nombre_reg FROM REGION'
        c.execute(query2)
        milista=[]
        for row in c.fetchall():
            milista.append(row[0])
            
        Label(frame, text = 'Región :').grid(row = 3, column = 0) 
        self.region = ttk.Combobox(frame,textvariable=self.region,values=milista)
        self.region.focus()
        self.region.grid(row = 3, column =1) 
        self.region.current(0)
        c.close()
        
        Label(frame, text = 'Nombre alumno:').grid(row = 5, column = 0)       
        self.nom_A = Entry(frame,textvariable=self.nom_A, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.nom_A.focus()
        self.nom_A.grid(row = 5, column = 1) 
        
        Label(frame, text = 'edad:').grid(row = 6, column = 0)       
        self.edad = Entry(frame,textvariable=self.edad,fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.edad.focus()
        self.edad.grid(row = 6, column = 1) 
        
        Label(frame, text = 'Contraseña:').grid(row = 7, column = 0)       
        self.contraseña = Entry(frame, textvariable=self.contraseña,fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.contraseña.focus()
        self.contraseña.grid(row = 7, column = 1) 
        
        Label(frame, text = 'Correo:').grid(row = 9, column = 0)       
        self.correo = Entry(frame, textvariable=self.correo,fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.correo.focus()
        self.correo.grid(row = 9, column = 1) 
        
        
        ttk.Button(frame,text="Guardar",command=self.agregar1,style="C.TButton").grid(row = 13, columnspan = 2, sticky = W + E)
        ttk.Button(frame,text="Salir",command=self.exit_btn,style="C.TButton").grid(row = 14, columnspan = 2, sticky = W + E)
        
    def exit_btn(self):
        self.rest.destroy()
        self.rest.update()
 

    def agregar1(self):
        try:
            db = sqlite3.connect( 'db_math.db')
            c = db.cursor()
            rut=self.rut_A.get()
            nom=self.nom_A.get()
            edad=self.edad.get()
            cont=self.contraseña.get()
            co=self.correo.get()
            es=self.establ.get()
            com=self.comuna.get()
            reg=self.region.get()
            
          
            c.execute("INSERT INTO ALUMNO VALUES(?,?,?,?,?,?,?,?,?)",(rut,es,com,reg,nom,edad,cont,co,None))
            db.commit() #Guardamos los cambios al terminar el ciclo
            mb.showinfo(title = "Datos insertados", message = 'Registrado')
        
        except sqlite3.OperationalError as error:
            mb.showerror(title = "Error", message = "Rellene todos los campos")  
            
    def contraseña(self):
        
        self.cor = Toplevel()
        self.cor.title("Recuperar contraseña")
        self.cor.geometry('720x360+500+200')
        self.cor['bg'] = 'skyblue'
        
        frame = LabelFrame(self.cor, text = 'Ingrese sus datos')
        frame.grid(row = 0, column = 0, columnspan = 3, pady = 20)
        
        Label(frame,text= 'Ingrese su rut' ).grid(row=3,column = 1)
        self.rut = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.rut.focus()
        self.rut.grid(row = 3, column = 2) 
        
        ttk.Button(frame,text="siguiente",command=self.siguiente,style="C.TButton").grid(row = 5, columnspan = 2, sticky = W + E)
    
    def siguiente(self):
        db = sqlite3.connect( 'db_math.db')
        c = db.cursor()
        us = self.rut.get()
        c.execute('SELECT * FROM ALUMNO WHERE rut_alumn = ?', (us,))
        
        if c.fetchall():
            mb.showinfo(title = "Correo Correcto", message = "Usuario encontrado")
            self.cambio()
        else:
            mb.showerror(title = "Correo incorrecto", message = "Usuario no encontrado")  
        c.close()

        
    def cambio(self):
        self.cons = Toplevel()
        self.cons.title("Recuperar contraseña")
        self.cons.geometry('720x360+500+200')
        self.cons['bg'] = 'skyblue'
        
        frame = LabelFrame(self.cons, text = 'Ingrese sus datos')
        frame.grid(row = 0, column = 0, columnspan = 3, pady = 20)
        
        Label(frame,text= 'Ingrese su contraseña nueva' ).grid(row=3,column = 1)
        self.cont = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.cont.focus()
        self.cont.grid(row =3, column = 2) 
        
        ttk.Button(frame,text="Guardar constraseña nueva",command=self.update_con,style="C.TButton").grid(row = 4, columnspan = 2, sticky = W + E)

         
    def update_con(self):
        
        db = sqlite3.connect( 'db_math.db')
        c = db.cursor()
        contraseña= self.cont.get()
        
        c.execute("""UPDATE ALUMNO set contraseña = ? """,contraseña)
                 
        
        if c.fetchall():
            mb.showinfo(title = "Contraseña actualizada", message = "Contraseña actualizada")
          
        else:
            mb.showerror(title = "Error", message = "Las contraseñas no son iguales,por favor ingrese nuevamente")  
        c.close()  
        
    def listar(self):
        
        self.perf = Toplevel()
        self.perf.geometry('720x360+500+200')
        self.perf['bg'] = 'skyblue'
        self.perf.title("Tabla de datos Publica")
        db = sqlite3.connect( 'db_math.db')
        c = db.cursor()
        
        query='SELECT rut_alumn,nombre_reg,nombre_com,id_puntaje FROM ALUMNO'
        midata=c.execute(query)
        milista=[rows for rows in midata]
        
        self.treeview = ttk.Treeview(self.perf,height = 10, columns = 4)
        self.treeview.insert("", tk.END, text= milista)
        self.treeview.pack()
        self.treeview.grid(row = 4, column =4, columnspan = 2)
        self.treeview.heading('#0', text = 'Perfil', anchor = CENTER)
        c.close()
        
        
    def nivel(self):
        self.ni = Toplevel()
        self.ni.geometry('920x565+300+100')
        self.ni['bg'] = 'skyblue'
        self.ni.title("Preguntas")
        
        self.img = Image.open('suma.png')
        self.img = self.img.resize((300, 250), Image.ANTIALIAS) # Redimension (Alto, Ancho)
        self.img = ImageTk.PhotoImage(self.img)

        self.img1 = Image.open('menos.png')
        self.img1 = self.img1.resize((300, 250), Image.ANTIALIAS) # Redimension (Alto, Ancho)
        self.img1 = ImageTk.PhotoImage(self.img1)

        self.img2 = Image.open('por.png')
        self.img2 = self.img2.resize((300, 250), Image.ANTIALIAS) # Redimension (Alto, Ancho)
        self.img2 = ImageTk.PhotoImage(self.img2)

        self.img3 = Image.open('division.png')
        self.img3 = self.img3.resize((300, 250), Image.ANTIALIAS) # Redimension (Alto, Ancho)
        self.img3 = ImageTk.PhotoImage(self.img3)

        self.img4 = Image.open('form.png')
        self.img4 = self.img4.resize((300, 250), Image.ANTIALIAS) # Redimension (Alto, Ancho)
        self.img4 = ImageTk.PhotoImage(self.img4)

        self.img5 = Image.open('boton-de-inicio.png')
        self.img5 = self.img5.resize((300, 250), Image.ANTIALIAS) # Redimension (Alto, Ancho)
        self.img5 = ImageTk.PhotoImage(self.img5)
        
        self.img6 = Image.open('perfil.png')
        self.img6 = self.img6.resize((300, 250), Image.ANTIALIAS) # Redimension (Alto, Ancho)
        self.img6 = ImageTk.PhotoImage(self.img6)

        botonNuevo1 = ttk.Button(self.ni, image=self.img, text="Fácil", compound="top",command=self.repositorio,style="C.TButton")
        botonNuevo1.grid(row = 0, column = 0)

        botonNuevo2 = ttk.Button(self.ni, image=self.img1, text="Medio", compound="top",command=self.repositorio2,style="C.TButton")
        botonNuevo2.grid(row = 0, column = 6)

        botonNuevo3 = ttk.Button(self.ni, image=self.img2, text="Dificil", compound="top",command=self.repositorio3,style="C.TButton")
        botonNuevo3.grid(row = 5, column =0)

        botonNuevo5 = ttk.Button(self.ni, image=self.img4, text="Formularios", compound="top",command=self.vocabulario,style="C.TButton")
        botonNuevo5.grid(row = 5, column = 6)

        botonNuevo6 = ttk.Button(self.ni, image=self.img5,text="Inicio",compound="top",command=self.nivel,style="C.TButton")
        botonNuevo6.grid(row = 0, column = 12)
        
        botonNuevo7 = ttk.Button(self.ni, image=self.img6,text="Perfil",compound="top",command=self.listar,style="C.TButton")
        botonNuevo7.grid(row = 5, column = 12)
        
    def repositorio(self):
        self.perf = Toplevel()
        self.perf.geometry('730x730+500+50') 
        self.perf['bg'] = 'skyblue'
        self.perf.title("Preguntas de suma")       
        
        db = sqlite3.connect( 'db_math.db')
        c = db.cursor()
        
    
        container = ttk.Frame(self.perf)
        canvas = tk.Canvas(container,width= 700, height= 700)
        scrollbar = ttk.Scrollbar(container, orient="vertical", command=canvas.yview)
        frame = ttk.Frame(canvas)

        frame.bind(
            "<Configure>",
            lambda e: canvas.configure(
                scrollregion=canvas.bbox("all")
            )
        )

        canvas.create_window((0, 0), window=frame, anchor="nw")

        canvas.configure(yscrollcommand=scrollbar.set)
        
        query1='SELECT nombre_p FROM PREGUNTA where id_pregunta=1'
        c.execute(query1)
        milista1=[]
        for row in c.fetchall():
            milista1.append(row[0])
            
        Label(frame, text = milista1).pack()
        self.respuesta1 = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2).pack()        
        ttk.Button(frame,text="Responder", command=self.validar_pregunta1,style="C.TButton").pack()

        query2='SELECT nombre_p FROM PREGUNTA where id_pregunta=2'
        c.execute(query2)
        milista2=[]
        for row in c.fetchall():
            milista2.append(row[0])
        Label(frame, text = milista2).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query3='SELECT nombre_p FROM PREGUNTA where id_pregunta=3'
        c.execute(query3)
        milista3=[]
        for row in c.fetchall():
            milista3.append(row[0])
      
        Label(frame, text = milista3).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query4='SELECT nombre_p FROM PREGUNTA where id_pregunta=4'
        c.execute(query4)
        milista4=[]
        for row in c.fetchall():
            milista4.append(row[0])
        
        Label(frame, text = milista4).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query5='SELECT nombre_p FROM PREGUNTA where id_pregunta=5'
        c.execute(query5)
        milista5=[]
        for row in c.fetchall():
            milista5.append(row[0])
 
        Label(frame, text = milista5).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        
        query7='SELECT nombre_p FROM PREGUNTA where id_pregunta=6'
        c.execute(query7)
        milista7=[]
        for row in c.fetchall():
            milista7.append(row[0])
            
        Label(frame, text = milista7 ).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()


        query8='SELECT nombre_p FROM PREGUNTA where id_pregunta=7'
        c.execute(query8)
        milista8=[]
        for row in c.fetchall():
            milista8.append(row[0])
        Label(frame, text = milista8).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query9='SELECT nombre_p FROM PREGUNTA where id_pregunta=8'
        c.execute(query9)
        milista9=[]
        for row in c.fetchall():
            milista9.append(row[0])
    
        Label(frame, text = milista9).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query10='SELECT nombre_p FROM PREGUNTA where id_pregunta=9'
        c.execute(query10)
        milista10=[]
        for row in c.fetchall():
            milista10.append(row[0])

        Label(frame, text = milista10).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query11='SELECT nombre_p FROM PREGUNTA where id_pregunta=10'
        c.execute(query11)
        milista11=[]
        for row in c.fetchall():
            milista11.append(row[0])
  
        Label(frame, text = milista11).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query13='SELECT nombre_p FROM PREGUNTA where id_pregunta=11'
        c.execute(query13)
        milista13=[]
        for row in c.fetchall():
            milista13.append(row[0])
     
        Label(frame, text = milista13).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query14='SELECT nombre_p FROM PREGUNTA where id_pregunta=12'
        c.execute(query14)
        milista14=[]
        for row in c.fetchall():
            milista14.append(row[0])
 
        Label(frame, text = milista14).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query15='SELECT nombre_p FROM PREGUNTA where id_pregunta=13'
        c.execute(query15)
        milista15=[]
        for row in c.fetchall():
            milista15.append(row[0])
   
        Label(frame, text = milista15).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query16='SELECT nombre_p FROM PREGUNTA where id_pregunta=14'
        c.execute(query16)
        milista16=[]
        for row in c.fetchall():
            milista16.append(row[0])
     
        Label(frame, text = milista16).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query17='SELECT nombre_p FROM PREGUNTA where id_pregunta=15'
        c.execute(query17)
        milista17=[]
        for row in c.fetchall():
            milista17.append(row[0])
     
        Label(frame, text = milista17).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query19='SELECT nombre_p FROM PREGUNTA where id_pregunta=16'
        c.execute(query19)
        milista19=[]
        for row in c.fetchall():
            milista19.append(row[0])
  
        Label(frame, text = milista19).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query20='SELECT nombre_p FROM PREGUNTA where id_pregunta=17'
        c.execute(query20)
        milista20=[]
        for row in c.fetchall():
            milista20.append(row[0])
     
        Label(frame, text = milista20).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        

        query21='SELECT nombre_p FROM PREGUNTA where id_pregunta=18'
        c.execute(query21)
        milista21=[]
        for row in c.fetchall():
            milista21.append(row[0])
     
        Label(frame, text = milista21).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        

        query22='SELECT nombre_p FROM PREGUNTA where id_pregunta=19'
        c.execute(query22)
        milista22=[]
        for row in c.fetchall():
            milista22.append(row[0])
  
        Label(frame, text = milista22).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query23='SELECT nombre_p FROM PREGUNTA where id_pregunta=20'
        c.execute(query23)
        milista23=[]
        for row in c.fetchall():
            milista23.append(row[0])
  
        Label(frame, text = milista23).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        
        query24='SELECT nombre_p FROM PREGUNTA where id_pregunta=21'
        c.execute(query24)
        milista24=[]
        for row in c.fetchall():
            milista24.append(row[0])
  
        Label(frame, text = milista24).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query25='SELECT nombre_p FROM PREGUNTA where id_pregunta=22'
        c.execute(query25)
        milista25=[]
        for row in c.fetchall():
            milista25.append(row[0])
  
        Label(frame, text = milista25).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query26='SELECT nombre_p FROM PREGUNTA where id_pregunta=23'
        c.execute(query26)
        milista26=[]
        for row in c.fetchall():
            milista26.append(row[0])
      
        Label(frame, text = milista26).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query27='SELECT nombre_p FROM PREGUNTA where id_pregunta=24'
        c.execute(query27)
        milista27=[]
        for row in c.fetchall():
            milista27.append(row[0])
   
        Label(frame, text = milista27).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        

        query28='SELECT nombre_p FROM PREGUNTA where id_pregunta=25'
        c.execute(query28)
        milista28=[]
        for row in c.fetchall():
            milista28.append(row[0])
       
        Label(frame, text = milista28).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query29='SELECT nombre_p FROM PREGUNTA where id_pregunta=26'
        c.execute(query29)
        milista29=[]
        for row in c.fetchall():
            milista29.append(row[0])
      
        Label(frame, text = milista29).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query30='SELECT nombre_p FROM PREGUNTA where id_pregunta=27'
        c.execute(query30)
        milista30=[]
        for row in c.fetchall():
            milista30.append(row[0])
        Label(frame, text = milista30).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query31='SELECT nombre_p FROM PREGUNTA where id_pregunta=28'
        c.execute(query31)
        milista31=[]
        for row in c.fetchall():
            milista31.append(row[0])
   
        Label(frame, text = milista31).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query32='SELECT nombre_p FROM PREGUNTA where id_pregunta=29'
        c.execute(query32)
        milista32=[]
        for row in c.fetchall():
            milista32.append(row[0])
     
        Label(frame, text = milista32).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query33='SELECT nombre_p FROM PREGUNTA where id_pregunta=30'
        c.execute(query33)
        milista33=[]
        for row in c.fetchall():
            milista33.append(row[0])
     
        Label(frame, text = milista33).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query34='SELECT nombre_p FROM PREGUNTA where id_pregunta=31'
        c.execute(query34)
        milista34=[]
        for row in c.fetchall():
            milista34.append(row[0])
       
        Label(frame, text = milista34).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query35='SELECT nombre_p FROM PREGUNTA where id_pregunta=32'
        c.execute(query35)
        milista35=[]
        for row in c.fetchall():
            milista35.append(row[0])
     
        Label(frame, text = milista35).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query36='SELECT nombre_p FROM PREGUNTA where id_pregunta=33'
        c.execute(query36)
        milista36=[]
        for row in c.fetchall():
            milista36.append(row[0])
    
        Label(frame, text = milista36).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query37='SELECT nombre_p FROM PREGUNTA where id_pregunta=34'
        c.execute(query37)
        milista37=[]
        for row in c.fetchall():
            milista37.append(row[0])
      
        Label(frame, text = milista37).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
 
        query38='SELECT nombre_p FROM PREGUNTA where id_pregunta=35'
        c.execute(query38)
        milista38=[]
        for row in c.fetchall():
            milista38.append(row[0])
       
        Label(frame, text = milista38).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
  
        query39='SELECT nombre_p FROM PREGUNTA where id_pregunta=36'
        c.execute(query39)
        milista39=[]
        for row in c.fetchall():
            milista39.append(row[0])
        
        Label(frame, text = milista39).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
 
        query40='SELECT nombre_p FROM PREGUNTA where id_pregunta=37'
        c.execute(query40)
        milista40=[]
        for row in c.fetchall():
            milista40.append(row[0])
       
        Label(frame, text = milista40).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query41='SELECT nombre_p FROM PREGUNTA where id_pregunta=38'
        c.execute(query41)
        milista41=[]
        for row in c.fetchall():
            milista41.append(row[0])
      
        Label(frame, text = milista41).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query42='SELECT nombre_p FROM PREGUNTA where id_pregunta=39'
        c.execute(query42)
        milista42=[]
        for row in c.fetchall():
            milista42.append(row[0])
     
        Label(frame, text = milista42).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
   
        query43='SELECT nombre_p FROM PREGUNTA where id_pregunta=40'
        c.execute(query43)
        milista43=[]
        for row in c.fetchall():
            milista43.append(row[0])
        Label(frame, text = milista43).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query44='SELECT nombre_p FROM PREGUNTA where id_pregunta=41'
        c.execute(query44)
        milista44=[]
        for row in c.fetchall():
            milista44.append(row[0])
     
        Label(frame, text = milista44).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query45='SELECT nombre_p FROM PREGUNTA where id_pregunta=42'
        c.execute(query45)
        milista45=[]
        for row in c.fetchall():
            milista45.append(row[0])
 
        Label(frame, text = milista45).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query46='SELECT nombre_p FROM PREGUNTA where id_pregunta=111'
        c.execute(query46)
        milista46=[]
        for row in c.fetchall():
            milista46.append(row[0])
       
        Label(frame, text = milista46).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query47='SELECT nombre_p FROM PREGUNTA where id_pregunta=112'
        c.execute(query47)
        milista47=[]
        for row in c.fetchall():
            milista47.append(row[0])
       
        Label(frame, text = milista47).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query48='SELECT nombre_p FROM PREGUNTA where id_pregunta=113'
        c.execute(query48)
        milista48=[]
        for row in c.fetchall():
            milista48.append(row[0])
        
        Label(frame, text = milista48).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query49='SELECT nombre_p FROM PREGUNTA where id_pregunta=114'
        c.execute(query49)
        milista49=[]
        for row in c.fetchall():
            milista49.append(row[0])
     
        Label(frame, text = milista49).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query50='SELECT nombre_p FROM PREGUNTA where id_pregunta=115'
        c.execute(query50)
        milista50=[]
        for row in c.fetchall():
            milista50.append(row[0])
        
        Label(frame, text = milista50).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query51='SELECT nombre_p FROM PREGUNTA where id_pregunta=116'
        c.execute(query51)
        milista51=[]
        for row in c.fetchall():
            milista51.append(row[0])
      
        Label(frame, text = milista51).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query52='SELECT nombre_p FROM PREGUNTA where id_pregunta=117'
        c.execute(query52)
        milista52=[]
        for row in c.fetchall():
            milista52.append(row[0])
     
        Label(frame, text = milista52).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
 
        query53='SELECT nombre_p FROM PREGUNTA where id_pregunta=118'
        c.execute(query53)
        milista53=[]
        for row in c.fetchall():
            milista53.append(row[0])
    
        Label(frame, text = milista53).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()        
        
        query54='SELECT nombre_p FROM PREGUNTA where id_pregunta=119'
        c.execute(query54)
        milista54=[]
        for row in c.fetchall():
            milista54.append(row[0])

        Label(frame, text = milista54).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query55='SELECT nombre_p FROM PREGUNTA where id_pregunta=120'
        c.execute(query55)
        milista55=[]
        for row in c.fetchall():
            milista55.append(row[0])

        Label(frame, text = milista55).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
 
        query56='SELECT nombre_p FROM PREGUNTA where id_pregunta=121'
        c.execute(query56)
        milista56=[]
        for row in c.fetchall():
            milista56.append(row[0])
   
        Label(frame, text = milista56).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query57='SELECT nombre_p FROM PREGUNTA where id_pregunta=122'
        c.execute(query57)
        milista57=[]
        for row in c.fetchall():
            milista57.append(row[0])
 
        Label(frame, text = milista57).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query58='SELECT nombre_p FROM PREGUNTA where id_pregunta=123'
        c.execute(query58)
        milista58=[]
        for row in c.fetchall():
            milista58.append(row[0])
  
        Label(frame, text = milista58).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query59='SELECT nombre_p FROM PREGUNTA where id_pregunta=124'
        c.execute(query59)
        milista59=[]
        for row in c.fetchall():
            milista59.append(row[0])

        Label(frame, text = milista59).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query60='SELECT nombre_p FROM PREGUNTA where id_pregunta=125'
        c.execute(query60)
        milista60=[]
        for row in c.fetchall():
            milista60.append(row[0])
 
        Label(frame, text = milista60).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
       
        query61='SELECT nombre_p FROM PREGUNTA where id_pregunta=126'
        c.execute(query61)
        milista61=[]
        for row in c.fetchall():
            milista61.append(row[0])

        Label(frame, text = milista61).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
  
        query62='SELECT nombre_p FROM PREGUNTA where id_pregunta=127'
        c.execute(query62)
        milista62=[]
        for row in c.fetchall():
            milista62.append(row[0])

        Label(frame, text = milista62).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query63='SELECT nombre_p FROM PREGUNTA where id_pregunta=128'
        midata63=c.execute(query63)
        c.execute(query63)
        milista63=[]
        for row in c.fetchall():
            milista63.append(row[0])
    
        Label(frame, text = milista63).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query64='SELECT nombre_p FROM PREGUNTA where id_pregunta=129'
        c.execute(query64)
        milista64=[]
        for row in c.fetchall():
            milista64.append(row[0])
        Label(frame, text = milista64).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query65='SELECT nombre_p FROM PREGUNTA where id_pregunta=130'
        c.execute(query65)
        milista65=[]
        for row in c.fetchall():
            milista65.append(row[0])
     
        Label(frame, text = milista65).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        
        query66='SELECT nombre_p FROM PREGUNTA where id_pregunta=131'
        c.execute(query66)
        milista66=[]
        for row in c.fetchall():
            milista66.append(row[0])
 
        Label(frame, text = milista66).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        
        query67='SELECT nombre_p FROM PREGUNTA where id_pregunta=132'
        c.execute(query67)
        milista67=[]
        for row in c.fetchall():
            milista67.append(row[0])

        Label(frame, text = milista67).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        
        query69='SELECT nombre_p FROM PREGUNTA where id_pregunta=133'
        c.execute(query69)
        milista69=[]
        for row in c.fetchall():
            milista69.append(row[0])
        Label(frame, text = milista69).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query70='SELECT nombre_p FROM PREGUNTA where id_pregunta=134'
        c.execute(query70)
        milista70=[]
        for row in c.fetchall():
            milista70.append(row[0]) 
    
        Label(frame, text = milista70).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        
        query71='SELECT nombre_p FROM PREGUNTA where id_pregunta=135'
        c.execute(query71)
        milista71=[]
        for row in c.fetchall():
            milista71.append(row[0])
       
        Label(frame, text = milista71).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query72='SELECT nombre_p FROM PREGUNTA where id_pregunta=136'
        c.execute(query72)
        milista72=[]
        for row in c.fetchall():
            milista72.append(row[0])
       
        Label(frame, text = milista72).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query73='SELECT nombre_p FROM PREGUNTA where id_pregunta=137'
        c.execute(query73)
        milista73=[]
        for row in c.fetchall():
            milista73.append(row[0])
       
        Label(frame, text = milista73).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query74='SELECT nombre_p FROM PREGUNTA where id_pregunta=138'
        c.execute(query74)
        milista74=[]
        for row in c.fetchall():
            milista74.append(row[0])
      
        Label(frame, text = milista74).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        ttk.Button(frame,text="inicio", command=self.nivel,style="C.TButton").pack()     
        self.img20= Image.open('ayuda.png')
        self.img20 = self.img20.resize((100, 100), Image.ANTIALIAS) # Redimension (Alto, Ancho)
        self.img20 = ImageTk.PhotoImage(self.img20)
        botonNuevo1 = tk.Button(frame, image=self.img20, text="Ayuda", compound="top",command=self.vocabulario)
        botonNuevo1.pack() 
        container.pack()
        canvas.pack(side="left", fill="both", expand=True)
        scrollbar.pack(side="right", fill="y")
        
    def repositorio2(self):
        self.perf = Toplevel()
        self.perf.geometry('730x730+500+50') 
        self.perf['bg'] = 'skyblue'
        self.perf.title("Preguntas de suma")       
        
        db = sqlite3.connect( 'db_math.db')
        c = db.cursor()
        
    
        container = ttk.Frame(self.perf)
        canvas = tk.Canvas(container,width= 700, height= 700)
        scrollbar = ttk.Scrollbar(container, orient="vertical", command=canvas.yview)
        frame = ttk.Frame(canvas)

        frame.bind(
            "<Configure>",
            lambda e: canvas.configure(
                scrollregion=canvas.bbox("all")
            )
        )

        canvas.create_window((0, 0), window=frame, anchor="nw")

        canvas.configure(yscrollcommand=scrollbar.set)
        
        query='SELECT nombre_p FROM PREGUNTA where id_pregunta=43'
        c.execute(query)
        milista=[]
        for row in c.fetchall():
            milista.append(row[0])
        Label(frame, text = milista).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2).pack()        
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query2='SELECT nombre_p FROM PREGUNTA where id_pregunta=44'
        c.execute(query2)
        milista2=[]
        for row in c.fetchall():
            milista2.append(row[0])
      
        Label(frame, text = milista2).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query3='SELECT nombre_p FROM PREGUNTA where id_pregunta=45'
        c.execute(query3)
        milista3=[]
        for row in c.fetchall():
            milista3.append(row[0])
      
        Label(frame, text = milista3).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query4='SELECT nombre_p FROM PREGUNTA where id_pregunta=46'
        c.execute(query4)
        milista4=[]
        for row in c.fetchall():
            milista4.append(row[0])
        
        Label(frame, text = milista4).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query5='SELECT nombre_p FROM PREGUNTA where id_pregunta=47'
        c.execute(query5)
        milista5=[]
        for row in c.fetchall():
            milista5.append(row[0])
 
        Label(frame, text = milista5).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        
        query7='SELECT nombre_p FROM PREGUNTA where id_pregunta=48'
        c.execute(query7)
        milista7=[]
        for row in c.fetchall():
            milista7.append(row[0])
        Label(frame, text = milista7 ).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()


        query8='SELECT nombre_p FROM PREGUNTA where id_pregunta=49'
        c.execute(query8)
        milista8=[]
        for row in c.fetchall():
            milista8.append(row[0])        
        Label(frame, text = milista8).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query9='SELECT nombre_p FROM PREGUNTA where id_pregunta=50'
        c.execute(query9)
        milista9=[]
        for row in c.fetchall():
            milista9.append(row[0])
    
        Label(frame, text = milista9).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query10='SELECT nombre_p FROM PREGUNTA where id_pregunta=51'
        c.execute(query10)
        milista10=[]
        for row in c.fetchall():
            milista10.append(row[0])

        Label(frame, text = milista10).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query11='SELECT nombre_p FROM PREGUNTA where id_pregunta=52'
        c.execute(query11)
        milista11=[]
        for row in c.fetchall():
            milista11.append(row[0])
  
        Label(frame, text = milista11).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query13='SELECT nombre_p FROM PREGUNTA where id_pregunta=53'
        c.execute(query13)
        milista13=[]
        for row in c.fetchall():
            milista13.append(row[0])
     
        Label(frame, text = milista13).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query14='SELECT nombre_p FROM PREGUNTA where id_pregunta=54'
        c.execute(query14)
        milista14=[]
        for row in c.fetchall():
            milista14.append(row[0])
 
        Label(frame, text = milista14).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query15='SELECT nombre_p FROM PREGUNTA where id_pregunta=55'
        c.execute(query15)
        milista15=[]
        for row in c.fetchall():
            milista15.append(row[0])
   
        Label(frame, text = milista15).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query16='SELECT nombre_p FROM PREGUNTA where id_pregunta=56'
        c.execute(query16)
        milista16=[]
        for row in c.fetchall():
            milista16.append(row[0])
     
        Label(frame, text = milista16).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query17='SELECT nombre_p FROM PREGUNTA where id_pregunta=57'
        c.execute(query17)
        milista17=[]
        for row in c.fetchall():
            milista17.append(row[0])
     
        Label(frame, text = milista17).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query19='SELECT nombre_p FROM PREGUNTA where id_pregunta=58'
        c.execute(query19)
        milista19=[]
        for row in c.fetchall():
            milista19.append(row[0])
  
        Label(frame, text = milista19).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query20='SELECT nombre_p FROM PREGUNTA where id_pregunta=59'
        c.execute(query20)
        milista20=[]
        for row in c.fetchall():
            milista20.append(row[0])
     
        Label(frame, text = milista20).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        

        query21='SELECT nombre_p FROM PREGUNTA where id_pregunta=60'
        c.execute(query21)
        milista21=[]
        for row in c.fetchall():
            milista21.append(row[0])
     
        Label(frame, text = milista21).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        

        query22='SELECT nombre_p FROM PREGUNTA where id_pregunta=61'
        c.execute(query22)
        milista22=[]
        for row in c.fetchall():
            milista22.append(row[0])
  
        Label(frame, text = milista22).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query23='SELECT nombre_p FROM PREGUNTA where id_pregunta=62'
        c.execute(query23)
        milista23=[]
        for row in c.fetchall():
            milista23.append(row[0])
  
        Label(frame, text = milista).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        
        query24='SELECT nombre_p FROM PREGUNTA where id_pregunta=63'
        c.execute(query24)
        milista24=[]
        for row in c.fetchall():
            milista24.append(row[0])
  
        Label(frame, text = milista).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query25='SELECT nombre_p FROM PREGUNTA where id_pregunta=64'
        c.execute(query25)
        milista25=[]
        for row in c.fetchall():
            milista25.append(row[0])
  
        Label(frame, text = milista25).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query26='SELECT nombre_p FROM PREGUNTA where id_pregunta=65'
        midata26=c.execute(query26)
        milista26=[]
        for row in c.fetchall():
            milista26.append(row[0])
      
        Label(frame, text = milista26).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query27='SELECT nombre_p FROM PREGUNTA where id_pregunta=66'
        c.execute(query27)
        milista27=[]
        for row in c.fetchall():
            milista27.append(row[0])
   
        Label(frame, text = milista27).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        

        query28='SELECT nombre_p FROM PREGUNTA where id_pregunta=67'
        c.execute(query28)
        milista28=[]
        for row in c.fetchall():
            milista28.append(row[0])
       
        Label(frame, text = milista28).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query29='SELECT nombre_p FROM PREGUNTA where id_pregunta=68'
        c.execute(query29)
        milista29=[]
        for row in c.fetchall():
            milista29.append(row[0])
      
        Label(frame, text = milista29).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query30='SELECT nombre_p FROM PREGUNTA where id_pregunta=69'
        c.execute(query30)
        milista30=[]
        for row in c.fetchall():
            milista30.append(row[0])
      
        Label(frame, text = milista30).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query31='SELECT nombre_p FROM PREGUNTA where id_pregunta=70'
        c.execute(query31)
        milista31=[]
        for row in c.fetchall():
            milista31.append(row[0])
   
        Label(frame, text = milista31).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query32='SELECT nombre_p FROM PREGUNTA where id_pregunta=71'
        c.execute(query32)
        milista32=[]
        for row in c.fetchall():
            milista32.append(row[0])
     
        Label(frame, text = milista32).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query33='SELECT nombre_p FROM PREGUNTA where id_pregunta=72'
        c.execute(query33)
        milista33=[]
        for row in c.fetchall():
            milista33.append(row[0])
     
        Label(frame, text = milista33).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query34='SELECT nombre_p FROM PREGUNTA where id_pregunta=73'
        c.execute(query34)
        milista34=[]
        for row in c.fetchall():
            milista34.append(row[0])
       
        Label(frame, text = milista34).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query35='SELECT nombre_p FROM PREGUNTA where id_pregunta=74'
        c.execute(query35)
        milista35=[]
        for row in c.fetchall():
            milista35.append(row[0])
     
        Label(frame, text = milista35).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query36='SELECT nombre_p FROM PREGUNTA where id_pregunta=75'
        c.execute(query36)
        milista36=[]
        for row in c.fetchall():
            milista36.append(row[0])
    
        Label(frame, text = milista36).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query37='SELECT nombre_p FROM PREGUNTA where id_pregunta=76'
        c.execute(query37)
        milista37=[]
        for row in c.fetchall():
            milista37.append(row[0])
      
        Label(frame, text = milista37).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
 
        query38='SELECT nombre_p FROM PREGUNTA where id_pregunta=77'
        c.execute(query38)
        milista38=[]
        for row in c.fetchall():
            milista38.append(row[0])
       
        Label(frame, text = milista38).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
  
        query39='SELECT nombre_p FROM PREGUNTA where id_pregunta=78'
        c.execute(query39)
        milista39=[]
        for row in c.fetchall():
            milista39.append(row[0])
        
        Label(frame, text = milista39).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
 
        query40='SELECT nombre_p FROM PREGUNTA where id_pregunta=79'
        c.execute(query40)
        milista40=[]
        for row in c.fetchall():
            milista40.append(row[0])
        Label(frame, text = milista40).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query41='SELECT nombre_p FROM PREGUNTA where id_pregunta=80'
        c.execute(query41)
        milista41=[]
        for row in c.fetchall():
            milista41.append(row[0])      
        Label(frame, text = milista41).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query42='SELECT nombre_p FROM PREGUNTA where id_pregunta=97'
        c.execute(query42)
        milista42=[]
        for row in c.fetchall():
            milista42.append(row[0])     
        Label(frame, text = milista42).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
   
        query43='SELECT nombre_p FROM PREGUNTA where id_pregunta=98'
        c.execute(query43)
        milista43=[]
        for row in c.fetchall():
            milista43.append(row[0])
        Label(frame, text = milista43).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query44='SELECT nombre_p FROM PREGUNTA where id_pregunta=99'
        c.execute(query44)
        milista44=[]
        for row in c.fetchall():
            milista44.append(row[0])
     
        Label(frame, text = milista44).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query45='SELECT nombre_p FROM PREGUNTA where id_pregunta=100'
        c.execute(query45)
        milista45=[]
        for row in c.fetchall():
            milista45.append(row[0])
 
        Label(frame, text = milista45).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query46='SELECT nombre_p FROM PREGUNTA where id_pregunta=101'
        midata46=c.execute(query46)
        milista46=[]
        for row in c.fetchall():
            milista46.append(row[0])
       
        Label(frame, text = milista46).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query47='SELECT nombre_p FROM PREGUNTA where id_pregunta=102'
        c.execute(query47)
        milista47=[]
        for row in c.fetchall():
            milista47.append(row[0])
       
        Label(frame, text = milista47).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query48='SELECT nombre_p FROM PREGUNTA where id_pregunta=103'
        c.execute(query48)
        milista48=[]
        for row in c.fetchall():
            milista48.append(row[0])        
        Label(frame, text = milista48).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query49='SELECT nombre_p FROM PREGUNTA where id_pregunta=104'
        c.execute(query49)
        milista49=[]
        for row in c.fetchall():
            milista49.append(row[0])     
        Label(frame, text = milista49).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query50='SELECT nombre_p FROM PREGUNTA where id_pregunta=105'
        c.execute(query50)
        milista50=[]
        for row in c.fetchall():
            milista50.append(row[0])
        
        Label(frame, text = milista50).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query51='SELECT nombre_p FROM PREGUNTA where id_pregunta=106'
        c.execute(query51)
        milista51=[]
        for row in c.fetchall():
            milista51.append(row[0])
      
        Label(frame, text = milista51).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query52='SELECT nombre_p FROM PREGUNTA where id_pregunta=107'
        c.execute(query52)
        milista52=[]
        for row in c.fetchall():
            milista52.append(row[0])
     
        Label(frame, text = milista52).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
 
        query53='SELECT nombre_p FROM PREGUNTA where id_pregunta=108'
        c.execute(query53)
        milista53=[]
        for row in c.fetchall():
            milista53.append(row[0])
    
        Label(frame, text = milista53).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()        
        
        query54='SELECT nombre_p FROM PREGUNTA where id_pregunta=109'
        c.execute(query54)
        milista54=[]
        for row in c.fetchall():
            milista54.append(row[0])

        Label(frame, text = milista54).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query55='SELECT nombre_p FROM PREGUNTA where id_pregunta=110'
        c.execute(query55)
        milista55=[]
        for row in c.fetchall():
            milista55.append(row[0])

        Label(frame, text = milista55).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
 
        query56='SELECT nombre_p FROM PREGUNTA where id_pregunta=111'
        c.execute(query56)
        milista56=[]
        for row in c.fetchall():
            milista56.append(row[0])
   
        Label(frame, text = milista56).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query57='SELECT nombre_p FROM PREGUNTA where id_pregunta=112'
        c.execute(query57)
        milista57=[]
        for row in c.fetchall():
            milista57.append(row[0])
 
        Label(frame, text = milista57).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query58='SELECT nombre_p FROM PREGUNTA where id_pregunta=113'
        c.execute(query58)
        milista58=[]
        for row in c.fetchall():
            milista58.append(row[0])
  
        Label(frame, text = milista58).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query59='SELECT nombre_p FROM PREGUNTA where id_pregunta=114'
        c.execute(query59)
        milista59=[]
        for row in c.fetchall():
            milista59.append(row[0])

        Label(frame, text = milista59).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query60='SELECT nombre_p FROM PREGUNTA where id_pregunta=115'
        c.execute(query60)
        milista60=[]
        for row in c.fetchall():
            milista60.append(row[0])
 
        Label(frame, text = milista60).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
       
        query61='SELECT nombre_p FROM PREGUNTA where id_pregunta=116'
        c.execute(query61)
        milista61=[]
        for row in c.fetchall():
            milista61.append(row[0])

        Label(frame, text = milista61).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
  
        query62='SELECT nombre_p FROM PREGUNTA where id_pregunta=117'
        c.execute(query62)
        milista62=[]
        for row in c.fetchall():
            milista62.append(row[0])

        Label(frame, text = milista62).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query63='SELECT nombre_p FROM PREGUNTA where id_pregunta=118'
        c.execute(query63)
        milista63=[]
        for row in c.fetchall():
            milista63.append(row[0])
    
        Label(frame, text = milista63).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query64='SELECT nombre_p FROM PREGUNTA where id_pregunta=119'
        c.execute(query64)
        milista64=[]
        for row in c.fetchall():
            milista64.append(row[0])
   
        Label(frame, text = milista64).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query65='SELECT nombre_p FROM PREGUNTA where id_pregunta=120'
        midata65=c.execute(query65)
        milista65=[]
        for row in c.fetchall():
            milista65.append(row[0])
     
        Label(frame, text = milista65).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        
        query66='SELECT nombre_p FROM PREGUNTA where id_pregunta=121'
        c.execute(query66)
        milista66=[]
        for row in c.fetchall():
            milista66.append(row[0])
 
        Label(frame, text = milista66).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        
        query67='SELECT nombre_p FROM PREGUNTA where id_pregunta=122'
        c.execute(query67)
        milista67=[]
        for row in c.fetchall():
            milista67.append(row[0])

        Label(frame, text = milista67).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        
        query69='SELECT nombre_p FROM PREGUNTA where id_pregunta=123'
        c.execute(query69)
        milista69=[]
        for row in c.fetchall():
            milista69.append(row[0])

        Label(frame, text = milista69).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query70='SELECT nombre_p FROM PREGUNTA where id_pregunta=124'
        c.execute(query70)
        milista70=[]
        for row in c.fetchall():
            milista70.append(row[0])
    
        Label(frame, text = milista70).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        
        query71='SELECT nombre_p FROM PREGUNTA where id_pregunta=125'
        midata71=c.execute(query71)
        milista71=[]
        for row in c.fetchall():
            milista71.append(row[0])
        Label(frame, text = milista71).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        ttk.Button(frame,text="inicio", command=self.nivel,style="C.TButton").pack()     
        self.img20= Image.open('ayuda.png')
        self.img20 = self.img20.resize((100, 100), Image.ANTIALIAS) # Redimension (Alto, Ancho)
        self.img20 = ImageTk.PhotoImage(self.img20)
        botonNuevo1 = tk.Button(frame, image=self.img20, text="Ayuda", compound="top",command=self.vocabulario)
        botonNuevo1.pack() 
        container.pack()
        canvas.pack(side="left", fill="both", expand=True)
        scrollbar.pack(side="right", fill="y")
        
    def repositorio3(self):
        self.perf = Toplevel()
        self.perf.geometry('730x730+500+50') 
        self.perf['bg'] = 'skyblue'
        self.perf.title("Preguntas de suma")       
        
        db = sqlite3.connect( 'db_math.db')
        c = db.cursor()
        
    
        container = ttk.Frame(self.perf)
        canvas = tk.Canvas(container,width= 700, height= 700)
        scrollbar = ttk.Scrollbar(container, orient="vertical", command=canvas.yview)
        frame = ttk.Frame(canvas)

        frame.bind(
            "<Configure>",
            lambda e: canvas.configure(
                scrollregion=canvas.bbox("all")
            )
        )

        canvas.create_window((0, 0), window=frame, anchor="nw")

        canvas.configure(yscrollcommand=scrollbar.set)
        query='SELECT nombre_p FROM PREGUNTA where id_pregunta=81'
        c.execute(query)
        milista=[]
        for row in c.fetchall():
            milista.append(row[0])
            
        Label(frame, text = milista).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2).pack()        
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query2='SELECT nombre_p FROM PREGUNTA where id_pregunta=82'
        c.execute(query2)
        milista2=[]
        for row in c.fetchall():
            milista2.append(row[0])
      
        Label(frame, text = milista2).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query3='SELECT nombre_p FROM PREGUNTA where id_pregunta=83'
        c.execute(query3)
        milista3=[]
        for row in c.fetchall():
            milista3.append(row[0])
      
        Label(frame, text = milista3).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query4='SELECT nombre_p FROM PREGUNTA where id_pregunta=84'
        c.execute(query4)
        milista4=[]
        for row in c.fetchall():
            milista4.append(row[0])
        
        Label(frame, text = milista4).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query5='SELECT nombre_p FROM PREGUNTA where id_pregunta=85'
        c.execute(query5)
        milista5=[]
        for row in c.fetchall():
            milista5.append(row[0])
        Label(frame, text = milista5).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        
        query7='SELECT nombre_p FROM PREGUNTA where id_pregunta=86'
        c.execute(query7)
        milista7=[]
        for row in c.fetchall():
            milista7.append(row[0])
        Label(frame, text = milista7 ).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()


        query8='SELECT nombre_p FROM PREGUNTA where id_pregunta=87'
        c.execute(query8)
        milista8=[]
        for row in c.fetchall():
            milista8.append(row[0])
        Label(frame, text = milista8).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query9='SELECT nombre_p FROM PREGUNTA where id_pregunta=88'
        c.execute(query9)
        milista9=[]
        for row in c.fetchall():
            milista9.append(row[0])
    
        Label(frame, text = milista9).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query10='SELECT nombre_p FROM PREGUNTA where id_pregunta=89'
        c.execute(query10)
        milista10=[]
        for row in c.fetchall():
            milista10.append(row[0])

        Label(frame, text = milista10).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query11='SELECT nombre_p FROM PREGUNTA where id_pregunta=90'
        c.execute(query11)
        milista11=[]
        for row in c.fetchall():
            milista11.append(row[0])
  
        Label(frame, text = milista11).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query13='SELECT nombre_p FROM PREGUNTA where id_pregunta=91'
        c.execute(query13)
        milista13=[]
        for row in c.fetchall():
            milista13.append(row[0])
     
        Label(frame, text = milista13).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query14='SELECT nombre_p FROM PREGUNTA where id_pregunta=92'
        c.execute(query14)
        milista14=[]
        for row in c.fetchall():
            milista14.append(row[0])
 
        Label(frame, text = milista14).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query15='SELECT nombre_p FROM PREGUNTA where id_pregunta=93'
        c.execute(query15)
        milista15=[]
        for row in c.fetchall():
            milista15.append(row[0])
   
        Label(frame, text = milista15).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query16='SELECT nombre_p FROM PREGUNTA where id_pregunta=94'
        c.execute(query16)
        milista16=[]
        for row in c.fetchall():
            milista16.append(row[0])
     
        Label(frame, text = milista16).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query17='SELECT nombre_p FROM PREGUNTA where id_pregunta=95'
        c.execute(query17)
        milista17=[]
        for row in c.fetchall():
            milista17.append(row[0])
     
        Label(frame, text = milista17).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query19='SELECT nombre_p FROM PREGUNTA where id_pregunta=96'
        c.execute(query19)
        milista19=[]
        for row in c.fetchall():
            milista19.append(row[0])
  
        Label(frame, text = milista19).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query20='SELECT nombre_p FROM PREGUNTA where id_pregunta=139'
        c.execute(query20)
        milista20=[]
        for row in c.fetchall():
            milista20.append(row[0])
     
        Label(frame, text = milista2).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        

        query21='SELECT nombre_p FROM PREGUNTA where id_pregunta=140'
        c.execute(query21)
        milista21=[]
        for row in c.fetchall():
            milista21.append(row[0])
     
        Label(frame, text = milista21).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        

        query22='SELECT nombre_p FROM PREGUNTA where id_pregunta=141'
        c.execute(query22)
        milista22=[]
        for row in c.fetchall():
            milista22.append(row[0])
  
        Label(frame, text = milista22).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        query23='SELECT nombre_p FROM PREGUNTA where id_pregunta=142'
        c.execute(query23)
        milista23=[]
        for row in c.fetchall():
            milista23.append(row[0])
  
        Label(frame, text = milista23).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()
        
        
        query24='SELECT nombre_p FROM PREGUNTA where id_pregunta=143'
        c.execute(query24)
        milista24=[]
        for row in c.fetchall():
            milista24.append(row[0])
  
        Label(frame, text = milista).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        query25='SELECT nombre_p FROM PREGUNTA where id_pregunta=144'
        c.execute(query25)
        milista25=[]
        for row in c.fetchall():
            milista25.append(row[0])
  
        Label(frame, text = milista25).pack()
        self.respuesta = Entry(frame, fg="#1034e8", font=("bold", 11), highlightthickness=2)
        self.respuesta.pack()
        ttk.Button(frame,text="Responder", command=self.validar_pregunta,style="C.TButton").pack()

        ttk.Button(frame,text="inicio", command=self.nivel,style="C.TButton").pack()     
        self.img20= Image.open('ayuda.png')
        self.img20 = self.img20.resize((100, 100), Image.ANTIALIAS) # Redimension (Alto, Ancho)
        self.img20 = ImageTk.PhotoImage(self.img20)
        botonNuevo1 = tk.Button(frame, image=self.img20, text="Ayuda", compound="top",command=self.vocabulario)
        botonNuevo1.pack() 
        container.pack()
        canvas.pack(side="left", fill="both", expand=True)
        scrollbar.pack(side="right", fill="y")   
        
    def vocabulario(self):
        self.perf = Toplevel()
        self.perf.geometry('720x360+500+50') 
        self.perf['bg'] = 'skyblue'
        self.perf.title("Vocabulario")
        file_name = "Presentation1.pdf"
        doc = fitz.open(file_name)

        zoom = 1
        mat = fitz.Matrix(zoom, zoom)

       #contar el número de páginas
        num_pages = 0
        for p in doc:
            num_pages += 1


    # agregar barra deslizante
        scrollbar = Scrollbar(self.perf)
        scrollbar.pack(side = RIGHT, fill = Y)
 
    # agregar canvas
        canvas = Canvas(self.perf, yscrollcommand = scrollbar.set)
        canvas.pack(side = LEFT, fill = BOTH, expand = 1)

    # definir el punto de entrada (campo para tomar entradas)
        entry = Entry(self.perf)

    # agregar label para los puntos de entrada
        label = Label(self.perf, text="Enter page number to display:")

        def pdf_to_img(page_num):
            page = doc.load_page(page_num)
            pix = page.get_pixmap(matrix=mat)
            return Image.frombytes("RGB", [pix.width, pix.height], pix.samples)

        def show_image():
            try:
                page_num = int(entry.get()) - 1
                assert page_num >= 0 and page_num < num_pages
                im = pdf_to_img(page_num)
                img_tk = ImageTk.PhotoImage(im)
                frame = Frame(canvas)
                panel = Label(frame, image=img_tk)
                panel.pack(side="bottom", fill="both", expand="yes")
                frame.image = img_tk
                canvas.create_window(0, 0, anchor='nw', window=frame)
                frame.update_idletasks()
                canvas.config(scrollregion=canvas.bbox("all"))
            except:
                pass
        #agregar botón para cambiar páginas
        button = Button(self.perf, text="Show Page", command=show_image)

        # poner locaciones visuales
        label.pack(side=TOP, fill=None)
        entry.pack(side=TOP, fill=BOTH)
        button.pack(side=TOP, fill=None)

        entry.insert(0, '1')
        show_image()

        scrollbar.config(command = canvas.yview)
        doc.close()

    def validar_pregunta1(self):
        
        db = sqlite3.connect( 'db_math.db')
        c = db.cursor()
        respuesta = self.respuesta.get()
        c.execute('SELECT respuesta FROM PREGUNTA WHERE id_pregunta = 1 AND respuesta = ? ', (respuesta, ))
        if c.fetchall():
            mb.showinfo(title = "Respuesta Correcta", message = "Respuesta correcta continue")
        else:
            mb.showerror(title = "Respuesta incorrecta", message = "Intente Otra vez")  
        c.close()        
    def validar_pregunta(self):
        
        db = sqlite3.connect( 'db_math.db')
        c = db.cursor()
        respuesta = self.respuesta.get()
        c.execute('SELECT respuesta FROM PREGUNTA WHERE id_pregunta = 1 AND respuesta = ? ', (respuesta, ))
        if c.fetchall():
            mb.showinfo(title = "Respuesta Correcta", message = "Respuesta correcta continue")
        else:
            mb.showerror(title = "Respuesta incorrecta", message = "Intente Otra vez")  
        c.close()        
          
        
        
if __name__ == '__main__':
    res = Tk()
    application = sesion(res)
    res.geometry('800x360+500+200') 
    res.resizable(0, 0)
    
    res.mainloop() 
 
```
